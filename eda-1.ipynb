```python
import sqlite3


conn = sqlite3.connect("house_prices.db")
```

```python
df.to_sql(
    name="houses",
    con=conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT SalePrice, OverallQual
FROM houses
LIMIT 5
"""

result = pd.read_sql(query, conn)

result
```

```python
pd.read_sql("PRAGMA table_info(houses);", conn)
```

```python
pd.read_sql("""
SELECT AVG(SalePrice) AS avg_price
FROM houses
""", conn)
```

```python
pd.read_sql("""
SELECT SalePrice, Neighborhood
FROM houses
ORDER BY SalePrice DESC
LIMIT 10
""", conn)
```

```python
pd.read_sql("""
SELECT Neighborhood,
       AVG(SalePrice) AS avg_price
FROM houses
GROUP BY Neighborhood
ORDER BY avg_price DESC
""", conn)
```
